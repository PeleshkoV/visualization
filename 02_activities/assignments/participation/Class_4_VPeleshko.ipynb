```python
# load necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# Make sample data
x = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1 = np.array([110, 180, 240, 99, 220])
y2 = np.array([170, 100, 90, 120, 50])

# define the figure and axes
fig, ax = plt.subplots(figsize=(7, 3))

# create bar and plot to the same axes
ax.bar(x, y1,
color = "indigo")
ax.plot(x, y2, color = "red")
```

```python
# calculate standard deviation of our data
y2_sd = np.std(y2)

# make the plot again
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

# add error information
ax.errorbar(x, y2,
yerr = y2_sd, # make vertical error bars
fmt = "none", # marks the error
ecolor= "green", # modify the error bars
elinewidth= 4,
capsize = 6,
capthick= 4,
errorevery= 2) # interval for error bars 
```

### Add images to plots

```python
from PIL import Image 
import requests 
from io import BytesIO 
```

```python
#get filefrom URL
response = requests.get('https://upload.wikimedia.org/wikipedia/en/c/cb/Monkey_D_Luffy.png')
image_file = BytesIO(response.content)
image = Image.open(image_file)
```

```python
# make a basic plot
fig, ax = plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color = "red")

# make a container for a image on the plot
ax_image = fig.add_axes([0.1, # x coordinate 
0.11, # y coordinate 
0.15, # image width
0.35]) # image height)

# add the image
ax_image.imshow(image)
ax_image.axis('off')
```
