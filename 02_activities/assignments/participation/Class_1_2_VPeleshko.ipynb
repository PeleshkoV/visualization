```python
# load necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# make some sample data
np.random.seed(613)  # set random seed for reproducibility
x = np.arange(50)  # create an array x
y = np.random.randint(0, 100,50)  # create an array y
```

```python
# make a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y)
```

```python
# make a vertical bar chart
fig, ax = plt.subplots(figsize=(5, 3))
ax.bar(x,y)
```

```python
# make a horizontal bar chart
fig, ax = plt.subplots(figsize=(5, 3))
ax.barh(x,y)
```

```python
# make a plot chart
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)
```

```python
# make a histogram
fig, ax = plt.subplots(figsize=(5, 3))
ax.hist(y)
```

```python
# Add axis labels and title to the plot chart
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

ax.set_title('Total growth over time')
ax.set_ylabel('Total growth')
ax.set_xlabel('Years since start')
fig.tight_layout()
```

```python
# Modify fonts of labels and titles
# First, set font dictionaries
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

# create plot chart 
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y)

# add modified title and labels
ax.set_title('Total growth over time', fontdict=font1, loc = 'left')
ax.set_ylabel('Total growth', fontdict=font2)
ax.set_xlabel('Years since start', fontdict=font2)
fig.tight_layout()
```

```python
# modify data points: set markers and change their color
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter( x,
y,
marker='*', 
color = "red") 
```

```python
# modify lines between data points and style of the markers
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot( x,
y,
marker='*', 
color = "blue",
linestyle = '--', # modify style of lines
linewidth = 2, # modify width of lines
markeredgecolor = '#fa9359', # set color edge of the markers
markerfacecolor = '#fa9359')  # fill the markers with color

# add grid lines on the chart, specify their width, style and color
ax.grid(axis = 'y', color = "grey", linewidth = 1, linestyle = '-')
```

```python
# Trial of Visualization from Python Graph Gallery. Example 1
# Use codes from the Gallery
height = [3, 12, 5, 18, 45]
bars = ('A', 'B', 'C', 'D', 'E')

df = pd.DataFrame({'height': height,
                   'bars': bars})

color = ['lightblue', 'blue', 'purple', 'red', 'green']
edgecolor = ['black']

# Create bars
fig, ax = plt.subplots()
ax.bar(df['bars'], df['height'],
       color=color,
       edgecolor=edgecolor,
       linewidth=1)

# Show graph
plt.show()
```
