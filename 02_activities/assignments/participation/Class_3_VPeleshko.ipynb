```python
# load necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
```

```python
# make some sample data
np.random.seed(613)
x = np.arange(50)
y1 = np.random.randint(0, 100,50)
y2 = np.random.randint(0, 100,50)
```

```python
# make a line plot with both of variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
```

```python
# add a label argument and legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right')
```

```python
# change different elements of the legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1" )
ax.plot(x,y2, label = "Person 2" )
ax.legend(loc='lower right',
          frameon = True, #add frame around the legend
          fontsize = 12, #change font size
          ncol = 2, #specify number of columns
          shadow = True) # add shadow around frame
```

```python
# move the legend outside of the box
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2")
ax.legend(loc='upper left',
bbox_to_anchor =(1, 1))
```

```python
# add text annotations 
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!")
```

```python
# modify text
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.text(10, 95, "This value is important!",
        ha = 'center', #horizont alignment
        color = 'red',
        size = 20)
```

```python
# Position text
# First, make an empty set of axes
fig, ax = plt.subplots()
ax.axis([0, 10, 0, 10])

# Second, position annotation with transform function
ax.text(1, 5, ". Data:(1, 5)", transform=ax.transData )
ax.text(0.5, 0.1, ". Axes:(0.5, 0.1)", transform=ax.transAxes )
ax.text(0.2, 0.2, ". Figure:(0.2, 0.2)", transform=fig.transFigure )
```

```python
# annotation with arrow
# First, create scatter plot with legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# create arrow with annotation
ax.annotate('This is important!', xy=(10, 95), # coordinate of data point
            xytext=(20, 94), # coordinate of text
arrowprops=dict(facecolor='black', shrink=0.05)) # dictionary of parameters for arrow
```

```python
# Change the style of the arrow
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.annotate('This is important!', xy=(10, 95), 
            xytext=(20, 94), 
arrowprops=dict(facecolor='hotpink', arrowstyle = "wedge")) 
```

```python
# remove only labels
# First, create scatter plot with legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# remove 
ax.xaxis.set_major_formatter(plt.NullFormatter())
ax.yaxis.set_major_formatter(plt.NullFormatter())
```

```python
# # remove tick marks and labels
# First, create scatter plot with legend
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

# remove
ax.yaxis.set_major_locator(plt.NullLocator())
ax.xaxis.set_major_locator(plt.NullLocator())
```

```python
# limit tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MaxNLocator(3))
ax.yaxis.set_major_locator(plt.MaxNLocator(3))
```

```python
# interval for tick marks
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')

ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.yaxis.set_major_locator(plt.MultipleLocator(10))
```

```python
# rotate axis labels
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1")
ax.scatter(x,y2, label = "Person 2")
ax.legend(loc='lower right')
          
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=45, ha='right')
```

# Subplot

```python
# create data set
np.random.seed(613)
x1 = np.arange(50)
y1 = np.random.randint(0, 75,50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
# create 2 subplots, modifying previous code for plots
fig,(ax1, ax2) = plt.subplots(ncols=2, # define location of this plots
    nrows=1,
    figsize=(7, 3))
ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# Customizing plots
fig,(ax1, ax2) = plt.subplots(ncols=2, # define location of this plots
    nrows=1,
    figsize=(7, 3))
ax1.scatter(x1,y1,
            marker = '*',
            color = 'red')
ax2.bar(x2,y2,
        color = 'green')
ax1.set_title('Bright')
ax2.set_title('Dark')
```

```python
# Subplots without a grid arrangement
fig, someaxes = plt.subplot_mosaic([['ax1', 'ax3'],
['ax2', 'ax3']],
figsize=(7, 4),
layout = 'constrained')

someaxes["ax1"].scatter(x1,y1)
someaxes["ax2"].bar(x2,y2)
someaxes["ax3"].plot(x1,y1)

# add labels to the plots
someaxes["ax1"].set_xlabel('A Big Label',fontsize=18)
someaxes["ax2"].set_xlabel('Another Label',fontsize=18)
someaxes["ax3"].set_xlabel('Label 2: 2 Fast 2 Furious',fontsize=18 )
```
